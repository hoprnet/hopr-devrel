```python
#imported data from the period of 27/10-17/11
```

```python
df = pd.read_csv('logs_8212280380033896655_1027_1117.csv')
```

```python
#creating new datetime
```

```python
df['new_timestamp'] = pd.to_datetime(df['tick_timestamp'])
```

```python
#difference between the tick_times
```

```python
df["time_diff"] = df['new_timestamp'].diff()
```

```python
#first value making = to the second (NaN cause problem with changing data type)
```

```python
df['time_diff'][0] = df['time_diff'][1]
```

```python
#modify the format of difference 
```

```python
df['time_diff2'] = df['time_diff'].apply(lambda x: x.total_seconds())
```

```python
df.describe()
```

```python
import plotly.express as px
```

```python
import plotly.graph_objects as go
```

```python
#creating bins
```

```python
cut_labels_15 = ['10.1-10.2', '10.2-10.3', '10.3-10.4', '10.4-10.5', '10.5-15', '15-20', '20-25', '25-30', '30-35', '35-40', '40-45', '45-50', '50-55', '55-60', '60+']
cut_bins = [10.100, 10.200, 10.300, 10.400, 10.500, 15.000, 20.000, 25.000, 30.000, 35.000, 40.000, 45.000, 50.000, 55.000, 60.000, 200000]
df['bins_h'] = pd.cut(df['time_diff2'], bins=cut_bins, labels=cut_labels_15)
```

```python
df['bins_h'].value_counts()
```

```python
import plotly.express as px
```

```python
df2 = df.sort_values(by=['time_diff2'], ascending=True)
```

```python
fig = px.histogram(df2, x="bins_h", 
                title='Ticktime difference, inst_id:8212280380033896655, time:27/10-17/11',
                   labels={'bins_h':'bins of seconds'}, # can specify one label per df column
                   opacity=0.8, # represent bars with log scale
                   )
fig.show()
```

```python
s = pd.to_datetime(df['new_timestamp'])
df3 = s.groupby(s.dt.floor('d')).size().reset_index(name='count')
```

```python
fig = px.bar(df3, x='new_timestamp', y='count', 
             title='Number of logs by day')
fig.show()
```

```python
p = pd.to_datetime(df['new_timestamp'])
df4 = p.groupby(p.dt.floor('h')).size().reset_index(name='count')
```

```python
fig = px.bar(df4, x='new_timestamp', y='count', 
             title='Number of logs by hour')
fig.show()
```
